```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config IPCompleter.greedy=True
```

```python
df=pd.read_csv(r'C:\Users\chat2\OneDrive\Documents\Machine-Learning-A-Z-New\Machine Learning A-Z New\Part 2 - Regression\Section 6 - Polynomial Regression\Position_Salaries.csv')
```

```python
X = df.iloc[:,1:2].values
Y = df.iloc[:,-1].values
```

```python
X.shape
#X = df.iloc[:,1:2].values give matrix of X whereas df.iloc[:,1] gives (10,)
```

```python
#Fitting Simple linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
#Fitting Polynomial Regression in dataset
from sklearn.preprocessing import PolynomialFeatures
#changes degree depending upon the accuracy of the model
poly_reg= PolynomialFeatures(degree = 2)
#poly_reg= PolynomialFeatures(degree =4 )
#2 specfies quadratic
X_poly=poly_reg.fit_transform(X)
```

```python
lin_reg2= LinearRegression()
lin_reg2.fit(X_poly,Y)
```

```python
#Vizualizing the linear regression result
plt.scatter(X,Y,color="red")
plt.plot(X, lin_reg.predict(X),color="blue")
plt.title('Truth or bluff (Linear Curve)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X, lin_reg2.predict(X_poly),color="blue")
plt.title('Truth or bluff (Polynomial curve)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```
